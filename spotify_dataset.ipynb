```python
df.info()
```

Con el atrubuto .info nos damos cuenta el número de variables que tiene el data set, el número de datos y el tipo de variable que contienen.

```python
df.dtypes
```

Aquí nos podemos dar cuenta qué tipo de variables son cada una.

* Numéricas: int64, float64

* Categóricas: Object
* Booleanas: bool (true/false)

Toca separar las variables por categorías para una mejor manilupación de los datos

```python
num_vars = df.select_dtypes(include=["int64", "float64"]).columns
cat_vars = df.select_dtypes(include=["object"]).columns
bool_vars = df.select_dtypes(include=["bool"]).columns

num_vars, cat_vars, bool_vars
```

##Rango de valores (solo para variables numéricas)

```python
df[num_vars].describe().T[["min", "max"]]
```

##Media, mediana y desviación estándar

```python
stats = df[num_vars].agg(["mean", "median", "std"]).T
stats
```
